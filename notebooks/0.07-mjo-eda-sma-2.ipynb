```python
from pathlib import Path

path = "../data/raw/"
files = Path(path).glob('*.csv')
```

```python
from pandas.core.frame import DataFrame


sma_dataframes: list[DataFrame] = [pd.read_csv(filepath, sep=";", low_memory=False) for filepath in files]
```

```python
from collections import Counter

all_columns = [col for df in sma_dataframes for col in df.columns]

column_counts = Counter(all_columns)

print("Column name occurrences:")
for col, count in column_counts.items():
    print(f"{col}: {count}")
```

```python
concatenated_sma = pd.concat(sma_dataframes, axis=0, ignore_index=True)
```

```python
concatenated_sma
```

```python
threshold = len(concatenated_sma) * 0.99  # 1% of total rows
concatenated_sma_cleaned = concatenated_sma.dropna(axis=1, thresh=threshold)
concatenated_sma_cleaned = concatenated_sma_cleaned.dropna()
```

```python
concatenated_sma_cleaned["año"] = concatenated_sma_cleaned["año"].astype(int)
```

```python
concatenated_sma_cleaned["cantidad_toneladas"] = concatenated_sma_cleaned["cantidad_toneladas"].replace(r'\.',"", regex=True).replace(",",".", regex=True).astype(float)
```
